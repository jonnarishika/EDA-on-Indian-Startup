```python
city_start_year = startup.groupby(['City','Starting Year']).size().reset_index(name='CompanyCount')
```

```python
def format_amount(amount):
    if amount >= 1_000_000_000:
        return f'${amount/1_000_000_000:.1f}B'
    elif amount >= 1_000_000:
        return f'${amount/1_000_000:.1f}M'
    elif amount >= 1_000:
        return f'${amount/1_000:.1f}K'
    else:
        return f'${amount:.0f}'
```

```python
# Step 2: Split the range
startup[['Min Employees', 'Max Employees']] = startup['No. of Employees'].str.split('-', expand=True)

# Step 3: Convert to integers (after removing None values)
startup = startup.dropna(subset=['Min Employees', 'Max Employees'])  # Optional: Drop rows with no employee info
startup['Min Employees'] = startup['Min Employees'].astype(int)
startup['Max Employees'] = startup['Max Employees'].astype(int)

# Step 4: Calculate average
startup['avg employees'] = (startup['Min Employees'] + startup['Max Employees']) / 2
startup['avg employees'] = startup['avg employees'].round().astype(int)
```

Tier 1: Bengaluru, Mumbai, Gurugram, and Delhi dominate across all metrics
Tier 2: Hyderabad, Chennai, Pune, and Ahmedabad show moderate activity
Tier 3: Cities like Jaipur, Kolkata, Patna, and Surat demonstrate minimal startup presence

2. <b>  Success Metrics </b>:
* Bengaluru: Leads in number of startups and employment generation (155,000 employees)

3. <b> Growth Pattern </b>:
* Early pioneers in Delhi,Mumbai,Chennai and Hyderabad (pre-2000)
* First major wave during the dot-com boom (2000)
* Second significant growth during mobile tech adoption (2007-2010)
* Peak startup formation in 2015 followed by 2019

4. <b> Sectoral Concentration: </b> :
* E-commerce emerges as the dominant vertical across all major hubs
* Certain cities demonstrate clear specialization (Hyderabad in healthcare, Ahmedabad in finance)
* Bengaluru shows the most balanced multi-industry presence
